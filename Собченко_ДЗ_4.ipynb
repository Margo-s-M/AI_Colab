# Завдання 1
Виведіть основні числові характеристики для числових ознак

вибрати колонки int64 float64 include= дозволяє обрати тільки ті колонки які потрібні

````python
numeric_df = df.select_dtypes(include=[np.number])
numeric_df
````

````python
summary_stats = numeric_df.describe()
summary_stats
````

````python
summary_stats["median"] = numeric_df.median()
summary_stats["var"] = numeric_df.var()
summary_stats["missing_values"] = numeric_df.isna().sum()
summary_stats
````
